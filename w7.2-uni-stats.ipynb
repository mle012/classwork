```python
import pandas as pd
from scipy import stats
from functions.eda import eda
url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/soccer-players.csv"
df = pd.read_csv(url)
cols = eda(df)

cols.loc[["height_in_cm", "date_of_birth"]]
```

```python
ht = df["height_in_cm"]
sk = stats.skew(ht, nan_policy= "omit")
kurt = stats.kurtosis(ht, nan_policy= "omit")
print (f"sk = {sk}; kurt = {kurt}")
```

```python
#IQR

q1 = ht.quantile(0.25)
q3 = ht.quantile(0.75)
iqr = q3 - q1
iqr
```

```python
lower_b = q1 - (1.5 * iqr)
upper_b = q3 + (1.5 * iqr)
outs = ht.loc[(ht<lower_b) | (ht>upper_b)] 
outs.to_list()
```

```python
ht = ht.loc[ht>150]
sk = stats.skew(ht, nan_policy="omit")
kurt = stats.kurtosis(ht, nan_policy = "omit")
print(f"sk = {sk:.4f} kurt = {kurt:.4f}")
```
