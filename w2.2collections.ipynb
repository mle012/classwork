# Collections

- set
- tuple
- list
- dict(dictionary)

```python
# set
# mutable, unordered (no index), no duplicates, iterable

set_inferred = {1,2,3,2} # almost never do this
set_explicit = set((2,3,4,4))

print(f"set_inferred = {set_inferred}")
print(f"set_explicit = {set_explicit}")
```

```python
set_inferred.add(7) # add
print(f"add = {set_inferred}")
# remove
set_inferred.remove(3)
print(f"remove set inferred = {set_inferred}")
# get the difference between sets
diff = set_inferred.difference(set_explicit)
print(f"diff = {diff}")
# get Overlaps between sets
overlap = set_inferred.intersection(set_explicit)
print(f"overlap = {overlap}")
```

```python
# tuple
# immutable, ordered (index), duplicates, iterable
tuple_inferred = (1,2,3,3,5,6)
tuple_explicit = tuple([4,5,6,7,2])

print(f"tuple_inferred = {tuple_inferred}")
print(f"tuple_explicit = {tuple_explicit}")
```

```python
# index_notation
# [start:stop:step]
print(f"index 0 = {tuple_inferred[0]}")
print(f"last item = {tuple_inferred[-1]}")
print(f"get 2nd and 3rd items = {tuple_inferred[1:3]}")
print(f"all but the last three = {tuple_inferred[:-3]}")
print(f"We only want the last three = {tuple_inferred[-3:]}")
print(f"tuple inferred has {len(tuple_inferred)} elements")
```

```python
#tuple methods
count = tuple_inferred.count(3) # get number of value
idx = tuple_inferred.index(3) # get index of value
print(count)
print(idx)
```

```python
# list
# mutable, ordered (index), duplicates, iterable
list_inferred = [1,2,3,4,5,5,6,3]
list_explicit = list((4,5,6,1,1,3,2))

print(f"list_inferred = {list_inferred}")
print(f"list_explicit = {list_explicit}")   
```

```python
# list methods

list_inferred.append(7)# add
print(f"append = {list_inferred}")

list_inferred.remove(5) # remove 
print(f"remove = {list_inferred}")

#pop
removed = list_inferred.pop() # removes last item or last index
print(f"removed = {removed}")
print(f"after pop = {list_inferred}")

# change

# extend
```

```python
# changing
list_inferred[0] = "Awesome Index"
list_inferred
```

```python
# dict (dictionary)
# mutable, {key:value}, unordered (index by key)
# no duplicates keys, but duplicate values, iterable
dict_inferred = {
    "name": "John", 
    "age": 30,
}
dict_explicit = dict(city="New York", country="USA")
print(f"dict_inferred = {dict_inferred}")
print(f"dict_explicit = {dict_explicit}")
```
